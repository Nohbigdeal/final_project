# 네이버 카페 크롤러 (태진님)

+ 사용법
    - cafecode에 카페 네임 입력
    - board_url에 &search=1 을 제거한 하단 페이지1 링크 입력
    - homeurl에 카페 주소 입력 후 수동 로그인 (생략 가능)

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import selenium
from urllib.parse import  urlparse
import time
import lxml
from urllib.request import urlopen
from selenium.webdriver.common.keys import Keys
import numpy as np
import pandas as pd
import re
import pickle
from contextlib import suppress
import time
```

```python
# 네이버 카페 게시판의 게시물을 수집하는 크롤러(제목, 본문, 등록일)
```

```python
# 기본 정보

cafe_code = "appleiphone" #str
# url 입력시 뒤에 있는 &search=1 을 제거한다
board_url = 'https://cafe.naver.com/ArticleList.nhn?search.clubid=12789486&search.menuid=157&search.boardtype=L&search.totalCount=151&search.cafeId=12789486'
page_range = 35  #int\n, 원하는 페이지수만큼 입력
post_num = []
contents_list = []
```

```python
# 대상 게시판의 지정된 페이지만큼 글 번호만 추출 (리스트화)

driver = None
driver = webdriver.Chrome("C:/3pm/pythonwork/analysis/chromedriver/chromedriver")

for i in range(page_range):
    url = board_url + "&search.page=" + str(i+1)  # 주소 지정
        
    driver.get(url)
    time.sleep(1) # implicity_wait, 암묵적, 파싱 대기시간. / #time.sleep은 무조건 대기
    driver.switch_to.frame('cafe_main')
    
    html = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    a = soup.find_all('div', {'class': 'board-number'})
    
    for n in a:
        post_num.append(re.search("[0-9]{2,8}", str(n)).group())
    
driver.quit()
```

```python
print(">>>>>>>>>>", cafe_code, "<<<<<<<<<<\n", post_num) # 내용 확인
```

```python
# 대상 게시물의 제목, 본문, 날짜 수집

driver = None
driver = webdriver.Chrome("C:/3pm/pythonwork/analysis/chromedriver/chromedriver")

temp_title = []
temp_content = []
temp_date = []

home_url = "https://cafe.naver.com/appleiphone"
driver.get(home_url)
time.sleep(20) # 수동 로그인

print("크롤링 시작")
for i in post_num:
    url = "https://cafe.naver.com/" + cafe_code + "/" + i
    driver.get(url)
    time.sleep(1) 

    driver.switch_to.frame('cafe_main')

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    content = soup.find('div', class_='ArticleContentBox')    

    with suppress(AttributeError):   # 제목 없는 게시글
        temp_title.append(content.find('h3',class_='title_text').text.strip()) 
        temp_content.append(content.find("div", class_="article_viewer").text.strip())
        temp_date.append(content.find('div',class_='article_info').find('span',class_='date').text.strip())
print("크롤링 완료")       
driver.quit()        
```

```python
# 내용 확인
arr = np.vstack([np.array(temp_title), np.array(temp_content), np.array(temp_date)])
df = pd.DataFrame(arr).T
df
```

```python
df.to_csv("아이패드_아사모카페.csv")
```
