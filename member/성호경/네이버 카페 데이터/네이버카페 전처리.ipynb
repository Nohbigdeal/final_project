## 전처리 함수

```python
def processing(data):
    # 데이터 중복 제거
    data.reset_index(inplace = True, drop = True)
    data["1"] = data["1"].drop_duplicates()
    data.dropna(inplace = True)
    data.reset_index(inplace = True, drop = True)

    # 데이터 필터링
    ban = ["택배", "배송"] # 추가 가능
    ban_idx = []
    try:
        for keyword in ban:
            for i in range(len(data)):
                if data["1"][i].find(keyword) != -1:
                    data = data.drop(index=i)
    
                elif data["1"][i].find(keyword) == -1 :
                    pass   
            data.reset_index(inplace = True, drop = True)
            
    except Exception as ex :
        print(ex, "오류발생")
        pass
    return data

# 해당 행을 바로 삭제 or 인덱스 번호를 저장 후 삭제
```

```python
# 한글만 살리기
def clean_str(s):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')    #한글만 남겨놓고 나머지 삭제
    s = hangul.sub(r'',str(s))

    cp = re.compile("["u"\U00010000-\U0010FFFF""]+", flags=re.UNICODE)  #BMP characters 이외 이모티콘삭제
    s=cp.sub(r'',s)

    s = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', s) #특수기호 삭제
 
    return s.strip()
```

## processing() / clear_str() 적용

```python
##### processing()

c_ipad = processing(c_ipad)
c_iphone = processing(c_iphone)
c_watch = processing(c_watch)
c_device = processing(c_device)
```

```python
##### clear_str()

# 아이패드
for i in range(len(c_ipad)):
    c_ipad["1"][i] = clean_str(c_ipad["1"][i])
    c_ipad["0"][i] = clean_str(c_ipad["0"][i])
c_ipad

# 아이폰
for i in range(len(c_iphone)):
    c_iphone["1"][i] = clean_str(c_iphone["1"][i])
    c_iphone["0"][i] = clean_str(c_iphone["0"][i])
c_iphone

# 애플워치
for i in range(len(c_ipad)):
    c_watch["1"][i] = clean_str(c_watch["1"][i])
    c_watch["0"][i] = clean_str(c_watch["0"][i])
c_watch

# 기타 기기
for i in range(len(c_ipad)):
    c_device["1"][i] = clean_str(c_device["1"][i])
    c_device["0"][i] = clean_str(c_device["0"][i])
c_device
```

```python
##### 파일 저장

c_ipad.to_csv("네이버 카페/c_ipad.csv")
c_iphone.to_csv("네이버 카페/c_iphone.csv")
c_watch.to_csv("네이버 카페/c_watch.csv")
c_device.to_csv("네이버 카페/c_device.csv")
```

```python
##### 혹시모르니 파일 병합

apple_cafe_total = pd.concat([c_ipad, c_iphone, c_watch, c_device])
apple_cafe_total.reset_index(drop = True, inplace = True)
apple_cafe_total.to_csv("네이버 카페/apple_cafe_total.csv")
```
